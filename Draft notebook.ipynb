```python
# import utils and basic libraries
import os
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from utils.preprocessing import gen_sub, bicubic
from utils.loss import l1_loss, psnr, ssim
from utils.prediction import ensemble, unensemble, shuffle_last_axis, predict_tensor, predict_tensor_permute, savePredictions, savePredictionsPermut
from utils.network import RAMS
from utils.training import Trainer
from skimage import io
from zipfile import ZipFile
```

```python
#-------------
# General Settings
#-------------
PATH_DATASET = 'dataset_mq' # pre-processed dataset path
name_net = 'RAMS' # name of the network
LR_SIZE = 128 # pathces dimension
SCALE = 3 # upscale of the proba-v dataset is 3
HR_SIZE = LR_SIZE * SCALE # upscale of the dataset is 3
OVERLAP = 32 # overlap between patches
CLEAN_PATH_PX = 0.85 # percentage of clean pixels to accept a patch
band = 'RED' # choose the band for the training
checkpoint_dir = f'ckpt/{band}_{name_net}' # weights path
log_dir = 'logs' # tensorboard logs path
submission_dir = 'result_mq' # submission dir
name_zip = 'submission_RAMS.zip'
```

```python
#-------------
# Network Settings
#-------------
FILTERS = 32 # features map in the network
KERNEL_SIZE = 3 # convolutional kernel size dimension (either 3D and 2D)
CHANNELS = 9 # number of temporal steps
R = 8 # attention compression
N = 12 # number of residual feature attention blocks
lr = 1e-4 # learning rate (Nadam optimizer)
BATCH_SIZE = 32 # batch size
EPOCHS_N = 100 # number of epochs
```

```python
# load MQ data
X_test_blue = np.load(os.path.join(PATH_DATASET, f'X_BLUE_test.npy'))
X_test_green = np.load(os.path.join(PATH_DATASET, f'X_GREEN_test.npy'))
X_test_red = np.load(os.path.join(PATH_DATASET, f'X_RED_test.npy'))
X_test_nir = np.load(os.path.join(PATH_DATASET, f'X_NIR_test.npy'))
```

# Prediction

```python
# build rams network
rams_network = RAMS(scale=SCALE, filters=FILTERS, 
                 kernel_size=KERNEL_SIZE, channels=CHANNELS, r=R, N=N)
```

```python
# load weights from checkpoint_dir
checkpoint = tf.train.Checkpoint(step=tf.Variable(0),
                                psnr=tf.Variable(1.0),
                                model=rams_network)

checkpoint.restore(tf.train.latest_checkpoint(checkpoint_dir))
```

```python
# predict proba-v test set with RAMS
X_preds_nir = []

for index in tqdm(range(X_test_nir.shape[0])):
    X_preds_nir.append(predict_tensor(rams_network, X_test_nir[index:index+1]))
```

```python
# predict proba-v test set with RAMS
X_preds_red = []

for index in tqdm(range(X_test_red.shape[0])):
    X_preds_red.append(predict_tensor(rams_network, X_test_red[index:index+1]))
```

```python
# predict proba-v test set with RAMS
X_preds_green = []

for index in tqdm(range(X_test_green.shape[0])):
    X_preds_green.append(predict_tensor(rams_network, X_test_green[index:index+1]))
```

```python
# predict proba-v test set with RAMS
X_preds_blue = []

for index in tqdm(range(X_test_blue.shape[0])):
    X_preds_blue.append(predict_tensor(rams_network, X_test_blue[index:index+1]))
```

```python
savePredictions(X_preds_nir, 'NIR', submission_dir)
savePredictions(X_preds_red, 'RED', submission_dir)
savePredictions(X_preds_green, 'GREEN', submission_dir)
savePredictions(X_preds_blue, 'BLUE', submission_dir)
```
