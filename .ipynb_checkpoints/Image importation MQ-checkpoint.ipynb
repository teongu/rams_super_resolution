```python
import rasterio
from rasterio.crs import CRS
from rasterio import warp
from rasterio.windows import Window
from pyproj import Proj
from math import floor, ceil
```

```python
from PIL import Image
```

```python
def longlat2window(xmin, ymin, xmax, ymax, dataset):
    p = Proj(dataset.crs)
    t = dataset.transform
    col_min, row_min = ~t * (xmin, ymin)
    col_max, row_max = ~t * (xmax, ymax)
    return Window.from_slices(rows=(floor(row_max), ceil(row_min)),
                              cols=(floor(col_min), ceil(col_max)))
```

```python
row_min, row_max, col_min, col_max = 3750, 4050, 2350,2750
```
